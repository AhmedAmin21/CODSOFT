# Feature Engineering

```python
dataset['trans_date_trans_time'].dtype
```

```python
dataset['trans_date_trans_time'] = pd.to_datetime(dataset['trans_date_trans_time'])
```

```python
dataset['year'] = dataset['trans_date_trans_time'].dt.year
dataset['month'] = dataset['trans_date_trans_time'].dt.month
dataset['day'] = dataset['trans_date_trans_time'].dt.day
dataset['hour'] = dataset['trans_date_trans_time'].dt.hour
dataset['minute'] = dataset['trans_date_trans_time'].dt.minute

dataset = dataset.drop(columns=['trans_date_trans_time'])
```

```python
dataset.head()
```

```python
dataset.head()
```

```python
from datetime import datetime
today = datetime.today()
dataset['dob'] = pd.to_datetime(dataset['dob'])

dataset['age'] = today.year - dataset['dob'].dt.year

dataset['age'] -= ((today.month & today.day) < (dataset['dob'].dt.month & dataset['dob'].dt.day))

dataset = dataset.drop(columns=['dob'])
```

# Encoding Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['category'] = le.fit_transform(dataset['category'])
dataset['gender'] = le.fit_transform(dataset['gender'])
```

```python
dataset.head()
```

```python
dataset = dataset.drop(columns=['lat', 'long', 'merch_lat', 'merch_long'])
```

```python
dataset.head()
```

```python
dataset.is_fraud.value_counts()
```

# ImBalanced Data
# OverSampling using SMOTE

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

X = dataset.drop('is_fraud', axis=1)
y = dataset['is_fraud']

smote = SMOTE()
X_res, y_res = smote.fit_resample(X, y)
```

```python
y_res.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier(criterion='entropy', random_state=42)
DT.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
y_pred = DT.predict(X_test)
print(accuracy_score(y_test, y_pred))
print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 25, criterion= 'entropy', random_state = 42)
classifier.fit(X_train, y_train)
```

```python
y_pred_r = classifier.predict(X_test)
print(accuracy_score(y_test, y_pred_r))
print(precision_score(y_test, y_pred_r))
print(recall_score(y_test, y_pred_r))
print(f1_score(y_test, y_pred_r))
```

```python
cm = confusion_matrix(y_test, y_pred_r)
cm
```
